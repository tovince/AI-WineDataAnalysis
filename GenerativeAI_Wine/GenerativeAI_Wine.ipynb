```python
# This is an example of how to implement a LLM into a Data Analysis workflow
# Use AI to analyze the correlation and answer questions.

import openai
import pandas as pd 
from sklearn.datasets import load_wine
```

```python
data = load_wine()
wine_df = pd.DataFrame(data.data, columns=data.feature_names)
wine_df['target']= data.target  #add 'target' column, where 'target' would be the output variable that our model would try to predict based on the other features
```

```python
#use generative AI to recommend top 5 wines
#openai.api_key = 'sk-proj-rx...'
import os

# Load the .env file
from dotenv import load_dotenv
load_dotenv()

# Use the API key
openai.api_key = os.getenv('OPENAI_KEY')

def query_openai(question):
    response = openai.ChatCompletion.create(
        model='gpt-4',
        messages=[{'role':'system','content':"You are a business analyst working for a winery and trying to recommend the best wine to customers based on data."},
                {'role':'user','content':question}])
    return response['choices'][0]['message']['content'].strip()

context = f"The dataset contains features like alcohol content, malic_acid, ash, alcalinity of ash and more. The target describes the quality of the win, 0 being the lowest, and 2 being the highest.\
This correlation matrix describes the relationship between the variables: {wine_df.corr()}."
question = 'Given these correlations, what features might be the most indicative of a high quality wine?'
response = query_openai(context+" "+question)
```

```python
context = f"The dataset contains features like alcohol content, malic_acid, ash, alcalinity of ash and more. The target describes the quality of the win, 0 being the lowest, and 2 being the highest.\
This correlation matrix describes the relationship between the variables: {wine_df.corr()}."
question = f"Based on your knowledge of wine characteristics, the ratings in this dataframe: {wine_df}, and the correlations between the features. \
I'd like to make some recommendations to my friends on which wines to buy. Can you show me the index values from the dataframe, which are the top 5 options for people who like sweet and the top 5 for people who like less sweet wines. Pick only from the highest rated wine?"
response = query_openai(context+" "+question)
```

From the "Response" above (formatted for easier reading):

**Correlation with Sweetness:** Based on the dataset and the correlation matrix, sweetness of wine generally correlates with the sugar (residual sugar, glucose+fructose) content. Unfortunately, residual sugar is not available in the data provided, but we can use some other properties such as alcohol, malic_acid and color_intensity as indicators of sweetness.

**Fruitier Wines:** Fruitier wines (higher sugar content) tend to have a higher alcohol content because the yeast consumes the sugar and produces alcohol during the fermentation process. Malic Acid is one of the primary acids in fruit, especially in apples, lending to a sweet flavor profile. Similarly, color intensity might give us an indication of the sweetness of wines as darker wines are typically sweeter (as sweetness generally comes from the fruit).

**Note:** These are general assumptions, the real sweetness level of a wine would rely on residual sugar and acidity levels of a wine, which is not present in the dataset.

**Top 5 Recommendations for Sweet Wine Lovers:** Lets say a customers preference for sweet wine is indicated by values above the median for alcohol, malic_acid and color_intensity.




```python
# Firstly, lets filter and sort the wine data for wines that have a quality
# rating of 2 (highest rating) and are considered sweet:

sweet_wines = wine_df[(wine_df['target'] == 2) &
                      (wine_df['alcohol'] > wine_df['alcohol'].median()) &
                      (wine_df['malic_acid'] > wine_df['malic_acid'].median()) &
                      (wine_df['color_intensity'] > wine_df['color_intensity'].median())]

top_sweet_wines = sweet_wines.sort_values(by=['alcohol', 'malic_acid', 'color_intensity'], ascending=False).head(5)
```

```python
print('Top 5 wines for sweet wine lovers: ', top_sweet_wines.index.tolist())
```

```python
less_sweet_wines = wine_df[(wine_df['target'] == 2) &
                           (wine_df['alcohol'] <= wine_df['alcohol'].median()) &
                           (wine_df['malic_acid'] <= wine_df['malic_acid'].median()) &
                           (wine_df['color_intensity'] <= wine_df['color_intensity'].median())]

top_less_sweet_wines = less_sweet_wines.sort_values(by=['alcohol', 'malic_acid', 'color_intensity']).head(5)
```

```python
print('Top 5 wines for less sweet wine lovers: ', top_less_sweet_wines.index.tolist())
```
